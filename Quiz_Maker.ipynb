```python
import random

def read_questions(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        questions = []
        current_question = ""
        for line in lines:
            if line.strip():
                current_question += line
            else:
                questions.append(current_question.strip())
                current_question = ""
        if current_question:
            questions.append(current_question.strip())
    return questions

def generate_quiz(questions, num_questions):
    random.shuffle(questions)
    quiz = questions[:num_questions]
    return quiz

def grade_quiz(quiz, key):
    score = 0
    total_questions = len(quiz)
    incorrect_answers = []

    for question in quiz:
        user_answer = input(f"{question}\nYour Answer: ")
        found = False

        for answer in key:
            question_key, correct_answer = answer.split(":")
            question_key = question_key.strip()

            if question_key in question:
                found = True
                if user_answer.lower() == correct_answer.strip().lower():
                    score += 1
                else:
                    incorrect_answers.append((question, correct_answer, user_answer))
                break

        if not found:
            print(f"No answer key found for question: {question}")

    percentage = (score / total_questions) * 100
   
    if len(incorrect_answers) > 0:
        print("\nIncorrect answers:")
        for question, correct_answer, user_answer in incorrect_answers:
            print(f"Question: {question}")
            print(f"Your Answer: {user_answer}")
            print(f"Correct Answer: {correct_answer}\n")
    
    return score, percentage


# File paths
questions_file = 'questions.txt'
key_file = 'key.txt'

# Read questions from file
questions = read_questions(questions_file)

# Generate quiz
num_questions = int(input("Enter the number of questions for the quiz in range[1-10]: "))
if num_questions > len(questions):
    print("Not enough questions available. Adjusting the number of questions.")
    num_questions = len(questions)
    exit()

quiz = generate_quiz(questions, num_questions)

# Read key from file
key = []
with open(key_file, 'r') as file:
    lines = file.readlines()
    for line in lines:
        line = line.strip()
        if line:
            key.append(line)

# Grade the quiz
score, percentage = grade_quiz(quiz, key)

# Provide feedback to the student
print(f"\nYour score: {score}/{num_questions}")
print(f"Percentage: {percentage}%")
```
